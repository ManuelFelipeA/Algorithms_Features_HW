### Graficas de las funciones y sus Omegas

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = np.linspace(0,10,10)
y = x

plt.plot(x,y,"k", linewidth=1)
```

```python
x = np.linspace(0.001,10,100)
y = x*(np.log(x))
plt.plot(y,"k", linewidth=1)
```

```python
x = np.linspace(0,10,100)
y = x**2

plt.plot(y,"k", linewidth=1)
```
